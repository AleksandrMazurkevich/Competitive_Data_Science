## Полезные функции

### Привязка Google Drive
Для того, чтобы расширить внутреннее хранилище составляющее от 40 - 100GB в зависимости от ускорителя, можно привязать сессию к Google Drive:


```python
from google.colab import drive
drive.mount('/content/gdrive')
```

И после перехода в `/content/gdrive/MyDrive/` будет доступен весь диск. 
Эта функция колаба является одной из важных, так как позволяет сохраняет обученные чекпоинты модели или, например, использовать данные, не подгружая их каждый раз из других источников.

```python
%cd /content/gdrive/MyDrive/
!ls
```

### Использование локального хранилища для подгрузки
Скорость чтения данных с Google Drive гораздо меньше, чем с сессионного хранилища Колаба. Поэтому перед обучением можно скопировать данные с диска в Колаб (в диреткорию `/content/`) и в дальнейшем использовать их оттуда.
